### Import data

```python
import os
import re

%cd /root/../content/drive/MyDrive/McHacks/
for folder in ['train', 'val', 'test']:
    file_list = os.listdir(f'./{folder}/labels')
    for file in file_list:
        if not file.endswith('.txt'):
            continue
        with open(f'./{folder}/labels/{file}', 'r+') as f:
            lines = f.read()
            replaced = re.sub(r'0(?=\s\d+\.\d+){4}', '1', lines)
            replaced = re.sub(r'2(?=\s\d+\.\d+){4}', '0', lines)
            f.seek(0)
            f.write(replaced)
            f.truncate()
            f.close()

# %cd /root/../content/yolov5/McHacks/
# for folder in ['train', 'val', 'test']:
#     file_list = os.listdir(f'./{folder}/labels')
#     for file in file_list:
#         if not file.endswith('.txt'):
#             continue
#         with open(f'./{folder}/labels/{file}', 'r+') as f:
#             lines = f.read()
#             replaced = re.sub(r'0(?=\s\d+\.\d+){4}', '1', lines)
#             replaced = re.sub(r'2(?=\s\d+\.\d+){4}', '0', lines)
#             f.seek(0)
#             f.write(replaced)
#             f.truncate()
#             f.close()
```
